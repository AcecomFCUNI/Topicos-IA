## **Matplotlib**
-----
Librería que permite visualizar datos mediante gráficos, histogramas, espectros de potencia, gráficos de barras, gráficos de error, gráficos de dispersión, etc...
<center>
<img src="https://matplotlib.org/stable/_images/anatomy.png" height = "20%" width = "50%">

```python
import matplotlib.pyplot as plt
import numpy as np
```

### Gráfico Lineal: Plot

```python
def f(x):
    return np.exp(-x)
x = np.linspace(-1, 5, num=10)
plt.plot(x, f(x), 'sg')# - 'gs' - 'go-' - 'rs' - 'r^' - 'g--'
plt.show()
```

```python
t = np.arange(0., 5., 0.2)
plt.plot(t, t, 'r--', t, t**2, 'bs', t, t**3, 'g^')
plt.show()
```

### Estilo Orientado a objetos y Pyplot

**Orientado a objetos:**

```python
x = np.linspace(0, 2, 100)

fig, ax = plt.subplots()  # Crea una figure y un axes.
ax.plot(x, x, label='linear')
ax.plot(x, x**2, label='quadratic')
ax.plot(x, x**3, label='cubic')
ax.set_xlabel('x label')
ax.set_ylabel('y label')
ax.set_title("Simple Plot")
ax.grid()
ax.legend()
```

**Pyplot:**

```python
x = np.linspace(0, 2, 100)

plt.plot(x, x, label='linear')  # Plotea data implicita en el axes.
plt.plot(x, x**2, label='quadratic')
plt.plot(x, x**3, label='cubic')
plt.xlabel('x label')
plt.ylabel('y label')
plt.title("Simple Plot")
plt.legend()
plt.grid()
plt.show()
```

### Interactivo y No Interactivo

**Interactivo:** plt.ion(), plt.draw()

```python
plt.ion()
for i in range(3):
    plt.plot(np.random.rand(10))
    plt.draw()
```

**No Interactivo:** plt.ioff(), plt.show()

```python
plt.ioff()
for i in range(3):
    plt.plot(np.random.rand(10))
    plt.show()
```

### Gráfico de Barras: Bar y Barh

```python
labels = ['G1', 'G2', 'G3', 'G4', 'G5']
men_means = [20, 34, 30, 35, 27]
women_means = [25, 32, 34, 20, 25]
x = np.arange(len(labels))  # ubicacion en eje x
width = 0.35  # ancho de barras

plt.bar(x - width/2, men_means, width, label='Men')
plt.bar(x + width/2, women_means, width, label='Women')

# Labels en eje x
plt.xticks(x, labels=labels)
plt.legend()

plt.show()
```

### Gráfico de Dispersión: Scatter

```python
N = 100
x = np.random.randn(N)
y = np.random.randn(N)
#un parametro mas
s = 50*np.random.randn(N) #tamaño
c = 50*np.random.randn(N) #color
plt.scatter(x, y, s=s, c=c)
plt.colorbar()
plt.show()
```

### Elipses: patches-Ellipse

```python
from matplotlib.patches import Ellipse

NUM = 50

for i in range(NUM):
    e = Ellipse(xy=np.random.rand(2) * 10,    #posicion
                width=np.random.rand(),       #ancho
                height=np.random.rand(),      #alto
                angle=np.random.rand() * 360) #rotacion

    plt.gca().add_artist(e)
    e.set_facecolor(np.random.rand(3))

plt.xlim(0, 10)
plt.ylim(0, 10)

plt.show()
```

### Histograma: Hist

```python
mu, sigma = 100, 15
x = mu + sigma * np.random.randn(10000)

n, bins, patches = plt.hist(x, 50, density=True, facecolor='g', alpha=0.75)

#añadir texto
plt.text(60, .025, '$\mu=100,\ \sigma=15$')
plt.xlim(40, 160)
plt.ylim(0, 0.03)
plt.grid()
plt.show()
```

### Gráfico circular: Pie

```python
vals = np.array([[60., 32.], [37., 40.], [29., 10.]])

plt.pie(vals.sum(axis=1), radius=1, wedgeprops=dict(width=0.3, edgecolor='w'))
plt.pie(vals.flatten(), radius=1-0.3, autopct='%1.1f%%',  wedgeprops=dict(width=0.5, edgecolor='w'))

plt.show()
```

### Coordenadas Polares: Polar

```python
r = np.arange(0, 1, 0.01)
theta = 2 * np.pi * r

plt.polar(theta, r, 'g.')
plt.grid(True)
plt.show()
```

### Gráfico de contorno: Contour

```python

delta = 0.025
x = np.arange(-3.0, 3.0, delta)
y = np.arange(-2.0, 2.0, delta)
X, Y = np.meshgrid(x, y)
Z1 = np.exp(-X**2 - Y**2)
Z2 = np.exp(-(X - 1)**2 - (Y - 1)**2)
Z = (Z1 - Z2) * 2

fig, ax = plt.subplots()
CS = ax.contour(X, Y, Z)
ax.clabel(CS, inline=True, fontsize=10)
plt.show()
```

### Grafico de estructura alámbrica: Plot_wireframe

```python
from mpl_toolkits.mplot3d import axes3d

fig = plt.figure()
ax = fig.add_subplot(projection='3d')

X, Y, Z = axes3d.get_test_data(0.06)

ax.plot_wireframe(X, Y, Z, rstride=10, cstride=20)

plt.show()
```

### Gráfico 3D: Plot_surface

```python
from matplotlib import cm
from mpl_toolkits.mplot3d.axes3d import get_test_data

fig, ax = plt.subplots(subplot_kw={"projection": "3d"})

X = np.arange(-5, 5, 0.25)
Y = np.arange(-5, 5, 0.25)
X, Y = np.meshgrid(X, Y)
R = np.sqrt(X**2 + Y**2)
Z = np.sin(R)

surf = ax.plot_surface(X, Y, Z, cmap=cm.coolwarm, linewidth=0)

plt.colorbar(surf, shrink=0.5, aspect=5)
plt.show()
```

### Gráfico de error: Errobar

```python
ax = plt.figure().add_subplot(projection='3d')

t = np.arange(0, np.pi/2, 0.01)
x, y, z = 2*t, 3*t, np.cos(5*t)

estep = 20
i = np.arange(t.size)
zuplims = (i % estep == 0) & (i // estep % 3 == 0)
zlolims = (i % estep == 0) & (i // estep % 3 == 2)

ax.errorbar(x, y, z, 0.2, zuplims=zuplims, zlolims=zlolims, errorevery=estep)

ax.set_xlabel("X label")
ax.set_ylabel("Y label")
ax.set_zlabel("Z label")

plt.show()
```

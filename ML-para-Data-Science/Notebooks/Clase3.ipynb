La tarea m√°s habitual a la hora de trabajar con matplotlib es representar una funci√≥n. Lo que tendremos que hacer es definir un dominio y evaluarla en dicho dominio. Por ejemplo:

$$f(x) = e^{-x^2}$$

```python
# definici√≥n de funci√≥n
def f(x):
    return np.exp(-x ** 2)
```

Definimos el dominio con la funci√≥n `np.linspace`, que crea un vector de puntos equiespaciados:

```python
x = np.linspace(-1, 5, num=30)
```

```python
x
```

```python
plt.plot(x, f(x))
```

```python
# etiquetas, label="Funci√≥n f(x)")
plt.plot(x, f(x))
plt.xlabel("Eje $x$")
plt.ylabel("$f(x)$")
```

```python
plt.plot(x, f(x))
plt.xlabel("Eje $x$")
plt.ylabel("$f(x)$")
plt.legend("f(x)")
plt.title("Funci√≥n $f(x)$")
```

```python
"""
=================================
Box plots with custom fill colors
=================================

This plot illustrates how to create two types of box plots
(rectangular and notched), and how to fill them with custom
colors by accessing the properties of the artists of the
box plots. Additionally, the ``labels`` parameter is used to
provide x-tick labels for each sample.

A good general reference on boxplots and their history can be found
here: http://vita.had.co.nz/papers/boxplots.pdf

"""
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np

# Random test data
np.random.seed(123)
all_data = [np.random.normal(0, std, 100) for std in range(1, 4)]
all_data

np.shape(all_data)

fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(12, 9)) #Figsize, width=12 pulgadas, heigth=9 pulgadas

fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(12, 9)) #Figsize, width=12, heigth=9
# boxplot rectangular
bplot1 = axes[0].boxplot(all_data,
                         vert=True,   
                         patch_artist=True)   

bplot2 = axes[1].boxplot(all_data,
                         notch=True,  
                         vert=True,   
                         patch_artist=True)   

# los colores de la paleta
colors = ['pink', 'lightblue', 'lightgreen']
for bplot in (bplot1, bplot2):
    for patch, color in zip(bplot['boxes'], colors):
        patch.set_facecolor(color)

# adicionando nombres a las abscisas
plt.setp(axes, xticks=[y+1 for y in range(len(all_data))],
         xticklabels=['x1', 'x2', 'x3', 'x4'])

plt.show()data_2 = np.random.normal(80, 30, 200)
data_2
```

```python
import plotly.plotly as py
import plotly.graph_objs as go
import numpy as np

y0 = datos.estatura
y1 = datos.cadera

trace0 = go.Box(
    y=y0,
    name = 'Estatura',
    marker = dict(
        color = 'rgb(214, 12, 140)',
    )
)
trace1 = go.Box(
    y=y1,
    name = 'Cadera',
    marker = dict(
        color = 'rgb(0, 128, 128)',
    )
)
data = [trace0, trace1]
iplot(data)
```

```python
import plotly.plotly as py
import plotly.figure_factory as ff
import numpy as np
 
x1 = datos.estatura[datos.sexo=='Fem']
x2 = datos.estatura[datos.sexo=='Mas']

hist_data = [x1, x2]

group_labels = ['Fem', 'Mas']

fig = ff.create_distplot(hist_data, group_labels, bin_size = .2)

iplot(fig, filename='Distplot with Multiple Datasets')
```

```python
import plotly.plotly as py
import plotly.figure_factory as ff
import numpy as np

x1 = datos.estatura[datos.sexo=='Fem']
x2 = datos.estatura[datos.sexo=='Mas']

hist_data = [x1, x2]

group_labels = ['Fem', 'Mas']
colors = ['#333F44', '#37AA9C']

fig = ff.create_distplot(hist_data, group_labels, show_hist = False, colors = colors)

fig['layout'].update(title='Curvas y rug')

iplot(fig, filename='Curvas y rug')
```

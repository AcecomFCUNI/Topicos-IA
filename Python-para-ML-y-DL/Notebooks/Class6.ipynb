## **Simple decorator**

```python
def my_decorator(func):
    """
    Arguments: 
        -func: You must pass any fuction 
    Return:
        - print_init_format: Return a function that print messages and "func"
    """
    def print_init_format():
        print("=============Initialize parameters===========\n")
        func()
        print("===========================================\n")
    return print_init_format

def initializer_weight():
    """
    This function Initialize the Weight and finally print this.
        - W1: Tensorflow tensor with dimension (2, 3)
        - b1: Tensorflow tensor zeros with dimension (2, 1)
        - W2: Tensorflow tensor with dimension (2, 2)
        - b2: Tensorflow tensor zeros with dimension (2, 1)       
    """
    W1 = tf.Variable(np.random.randn(2, 3))*0.01
    b1 = tf.Variable(np.zeros((2, 1)))
    W2 = tf.Variable(np.random.randn(2, 2))*0.01
    b2 = tf.Variable(np.zeros((2, 1)))
    print("W1: {},\n\nW2: {},\n\nb1: {},\n\nb2: {} ".format(W1, W2, b1, b2))

decorator = my_decorator(initializer_weight) 
#Use the "initializer_weight" fuction in the "my_decorator" function

decorator()
```

### **[Pie-decorator](https://www.python.org/dev/peps/pep-0318/#background)**

Python allows you to use decorators in a simpler way with the @ symbol, sometimes called the “pie” syntax. 


```python
def my_decorator(function):
    """
    Arguments: 
        -func: You must pass any fuction 
    Return:
        - print_init_format: Return a function that print messages and "func"
    """
    def print_Init_format():
        print("=============Initialize parameters===========\n")
        function()
        print("===========================================\n")
    return print_Init_format

@my_decorator #Use Pie sintaxis
def initializer_weight():
    """
    This function Initialize the Weight and finally print this.
        - W1: Tensorflow tensor with dimension (2, 3)
        - b1: Tensorflow tensor zeros with dimension (2, 1)
        - W2: Tensorflow tensor with dimension (2, 2)
        - b2: Tensorflow tensor zeros with dimension (2, 1)       
    """
    W1 = tf.constant(np.random.randn(2, 3))*0.01
    b1 = tf.constant(np.zeros((2, 1)))
    W2 = tf.constant(np.random.randn(2, 3))*0.01
    b2 = tf.constant(np.zeros((2, 1)))

    print("W1: {},\n\nW2: {},\n\nb1: {},\n\nb2: {} ".format(W1, W2, b1, b2))

initializer_weight()
```

## **Decorating Functions With Arguments**


```python
def decorator(func):
    """
    Arguments: 
        -func: this is any fuction.
    Return:
        - operation: this fuction print and call the "func" 
    """
    def operation(*args, **kwargs):
        print("Initializer Operation",end='\n')
        result = func(*args, **kwargs)
        print(f"Operation:{result[1]}, Result: {result[0]}\n")
    return operation

@decorator
def suma(a, b, operation = 'suma'):
    """
    Arguments:
        - a: First argument type (int)
        - b: Second argument type (int)
        - operation: Name Operation (str)
    Return:
        - this fuction return sum on "a" and "b" and the operation.
    """
    return (a+b, operation)
@decorator
def resta(a, b, operation = 'resta'):
    """
    Arguments:
        - a: First argument type (int)
        - b: Second argument type (int)
        - operation: Name Operation (str)
    Return:
        - this fuction return sum on "a" and "b" and the operation.
    """    
    return (a-b, operation)

suma(4, 5)

resta(200, 31)
```

```python
def initilizer(func):
    """
    Arguments: 
        -func: this is any fuction.
    Return:
        - wrapper: this fuction call the "func" , finally print value and value > 0.5
    """
    def wrapper(*args):
        value = func(*args)
        print(
            f"Original sigmoid:\n\n{value},\n\nSigmoid > 0.5\n\n{value[value > 0.5]}"
        )

    return wrapper


@initilizer
def sigmoid(x):
    """
    Argument:
        - x: This is an argument.
    Return:
        - return: Return Sigmoid operation.
    """
    return 1 / (1 + np.exp(-x))


sigmoid(torch.tensor(np.random.randn(3, 2)))
```

### **Numba**

Numba is a just-in-time compiler that will allow us to speed up the python code and that is useful when using nested loops.

[Just-in-time](https://en.wikipedia.org/wiki/Just-in-time_compilation):

In computing, just-in-time (JIT) compilation (also dynamic translation or run-time compilations) is a way of executing computer code that involves compilation during execution of a program – at run time – rather than before execution.

```python
from numba import jit 
from time import time
```

```python
matrix_random = np.random.rand(100,100) #Initialize a matrix (100, 100)
```

```python
def multiplicacion_matricial(X, Y):
    """
    Argmuents:
        - X: This is a numpy matrix 
        - Y: This is a  numpy matrix
    Return:
        - Return the matrix multiplication
    """
    matrix_mult = np.zeros_like(matrix_random)
    for rown in range(len(X)):
        for  column in range(len(Y[0])):
            for aux in range(len(Y)):
                matrix_mult[rown, column] += X[rown, aux]*Y[aux, column]
    return matrix_mult
```

```python
start = time()

multiplicacion_matricial(matrix_random, matrix_random)

finish = time()

print(f"time: {finish - start} ")
```

```python
@jit
def multiplicacion_matricial_jit(X, Y):
    """
    Argmuents:
        - X: This is a numpy matrix 
        - Y: This is a  numpy matrix
    Return:
        - Return the matrix multiplication
    """
    matrix_mult = np.zeros_like(matrix_random)
    for rown in range(len(X)):
        for  column in range(len(Y[0])):
            for aux in range(len(Y)):
                matrix_mult[rown, column] += X[rown, aux]*Y[aux, column]
    return matrix_mult
```

```python
start = time()

multiplicacion_matricial_jit(matrix_random, matrix_random)

finish = time()

print(f"time: {finish - start} ")
```

**7/8 Zeta Riemann function**

```python
def serie(n):
    """
    Argument:
        - n : Number of elements in the sum
    Return:
        - sum : This is the result of the Zeta Riemman aprox
    """
    sum=0
    for i in range(1,n):
        sum=sum+1/(2*i-1)**3
    return sum
```

```python
start = time()
serie(1000)
finish = time()
print(f"time: {finish - start} ")
```

```python
@jit
def serie_numba(n):
    """
    Argument:
        - n : Number of elements in the sum
    Return:
        - sum : This is the result of the Zeta Riemman aprox
    """
    sum=0
    for i in range(1,n):
        sum=sum+1/(2*i-1)**3
    return sum
```

```python
start = time()
serie_numba(1000)
finish = time()
print(f"time: {finish - start} ")
```

```python
def L1_loss(y, y_hat):
    """
    Argument:
        - Y : Array with shape (n, n)
        - Y : Array with shape (n, n)
    Return:
        return the L1 loss 
    """
    return tf.reduce_mean(tf.abs(y - y_hat))
```
